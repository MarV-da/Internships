```python
dv1=pd.read_csv('Datos Ventas I.csv', header=3)
dv1
```

```python
dv1=dv1.drop(['Unnamed: 0','Unnamed: 6','Mes Nro.','Mes Año'], axis=1)
dv1
```

```python
dv1.info()
```

```python
dv1['Fecha']=dv1['Fecha'].astype('datetime64[ns]')
dv1['Vendedor']=dv1['Vendedor'].astype('string')
dv1['Mes']=dv1['Mes'].astype('string')
```

```python

dv1['Importe IVA Inc. (UYU)']=dv1['Importe IVA Inc. (UYU)'].str.replace('.','', regex=False)
dv1['Importe IVA Inc. (UYU)']=dv1['Importe IVA Inc. (UYU)'].str.replace(',','.', regex=False)
dv1['Importe IVA Inc. (UYU)']=dv1['Importe IVA Inc. (UYU)'].astype('float64')
```

```python
dv1.info()
```

```python
meses={'enero':1,'febrero':2,'marzo':3,'abril':4,'mayo':5,'junio':6,'julio':7,'agosto':8,'setiembre':9,'octubre':10,'noviembre':11,'diciembre':12}
```

```python
dv1['Mes_num']=dv1['Mes'].map(meses)
dv1
```

```python
dv1.info()
```

```python
mensual_total=dv1.groupby(['Mes_num','Mes','Año'],as_index=False)['Importe IVA Inc. (UYU)'].sum().sort_values(by=['Año','Mes_num'])
mensual_total
```

```python
mensual_total['var_mensual_total']=mensual_total.groupby(['Mes'])['Importe IVA Inc. (UYU)'].pct_change()
mensual_total.rename(columns={'Importe IVA Inc. (UYU)':'Importe IVA Inc. (UYU) Mensual Total'},inplace=True)
mensual_total
```

```python
anual_total=dv1.groupby(['Año'],as_index=False)['Importe IVA Inc. (UYU)'].sum().sort_values(by=['Año'])
anual_total
```

```python
anual_total['var_anual_total']=anual_total['Importe IVA Inc. (UYU)'].pct_change()
anual_total.rename(columns={'Importe IVA Inc. (UYU)':'Importe IVA Inc. (UYU) Anual Total'},inplace=True)
anual_total
```

```python
mensual_vendedor=dv1.groupby(['Año','Mes_num','Mes','Vendedor'],as_index=False)['Importe IVA Inc. (UYU)'].sum().sort_values(by=['Año','Mes_num'])
mensual_vendedor
```

```python
mensual_vendedor['var_mensual_vendedor']=mensual_vendedor.groupby(['Vendedor','Mes'])['Importe IVA Inc. (UYU)'].pct_change()
mensual_vendedor.rename(columns={'Importe IVA Inc. (UYU)':'Importe IVA Inc. (UYU) Vendedor Mensual'},inplace=True)
mensual_vendedor
```

```python
anual_vendedor=dv1.groupby(['Año','Vendedor'])['Importe IVA Inc. (UYU)'].sum().reset_index().sort_values(by=['Año'])
anual_vendedor
```

```python
anual_vendedor['var_anual_vendedor']=anual_vendedor.groupby(['Vendedor'])['Importe IVA Inc. (UYU)'].pct_change()
anual_vendedor.rename(columns={'Importe IVA Inc. (UYU)':'Importe IVA Inc. (UYU) Vendedor Anual'},inplace=True)    
anual_vendedor
```

```python
df_final=mensual_vendedor.merge(mensual_total,on=['Año','Mes_num','Mes'],how='left')
df_final=df_final.merge(anual_vendedor,on=['Año','Vendedor'],how='left')
df_final=df_final.merge(anual_total,on=['Año'],how='left')
df_final
```

```python
df_final.to_csv('Datos Ventas I Consolidado.csv',index=False)
```

```python
df_final_fig=px.bar(df_final, x='Año', y='Importe IVA Inc. (UYU) Mensual Total', barmode='group', color='Mes',hover_data={'var_mensual_total':':.2%'},title='Variación de Ventas Mensual por Año',labels={'Importe IVA Inc. (UYU) Mensual Total':'Ventas'})
df_final_fig
```

```python
df_final_fig2=px.bar(df_final, x='Año', y='Importe IVA Inc. (UYU) Anual Total',hover_data={'var_anual_total':':.2%'},title='Variación de Ventas Anual por Año',labels={'Importe IVA Inc. (UYU) Anual Total':'Ventas'})
df_final_fig2
```

```python
df_final_fig.update_traces(texttemplate='%{text:.2%}')
```
