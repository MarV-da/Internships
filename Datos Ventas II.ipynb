```python
dv2=pd.read_csv('Datos Ventas II.csv', header=3)
dv2
```

```python
dv2.columns
```

```python
dv2.drop(columns=['Unnamed: 0','NetoExento','NetoMinimo','NetoBasico','neto','IvaMinimo','IvaBasico','Imp_cli'], inplace=True)
dv2
```

```python
dv2.info()
```

```python
dv2['Total']=dv2['Total'].str.replace('.','',regex=False)
dv2['Total']=dv2['Total'].str.replace(',','.',regex=False)
dv2['Total']=dv2['Total'].astype('float64')
```

```python
dv2['Cliente'].astype('string')
dv2['RUT'].astype('string')
dv2['Mes'].astype('string')
dv2.info()
```

```python
ventas_anuales=dv2.groupby(['Cliente'],as_index=False)['Total'].sum().sort_values('Total',ascending=False)
ventas_anuales
```

```python
ventas_anuales.to_csv('Datos Ventas II Total.csv', index=False)
```

```python
ventas_anuales['Porcentaje']=ventas_anuales['Total']/ventas_anuales['Total'].sum()
ventas_anuales
```

```python
ventas_anuales['Porcentaje Acumulado']=ventas_anuales['Porcentaje'].cumsum()
ventas_anuales
```

```python
clientes_top=ventas_anuales[ventas_anuales['Porcentaje Acumulado']<=0.80]
clientes_top
```

```python
top_20=clientes_top.sort_values('Total',ascending=False)[:20]
top_20
```

```python
top_20_fig=px.bar(top_20, x='Cliente', y='Total', title='Top 20 Clientes por Ventas', labels={'Total':'Ventas'})
top_20_fig
```
