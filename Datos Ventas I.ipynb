```python
dv1=pd.read_csv('Datos Ventas I.csv', header=3)
dv1
```

```python
dv1=dv1.drop(['Unnamed: 0','Unnamed: 6','Mes Nro.','Mes Año'], axis=1)
dv1
```

```python
dv1.info()
```

```python
dv1['Fecha']=dv1['Fecha'].astype('datetime64[ns]')
dv1['Vendedor']=dv1['Vendedor'].astype('string')
dv1['Mes']=dv1['Mes'].astype('string')
```

```python

dv1['Importe IVA Inc. (UYU)']=dv1['Importe IVA Inc. (UYU)'].str.replace('.','', regex=False)
dv1['Importe IVA Inc. (UYU)']=dv1['Importe IVA Inc. (UYU)'].str.replace(',','.', regex=False)
dv1['Importe IVA Inc. (UYU)']=dv1['Importe IVA Inc. (UYU)'].astype('float64')
```

```python
dv1.info()
```

```python
meses={'enero':1,'febrero':2,'marzo':3,'abril':4,'mayo':5,'junio':6,'julio':7,'agosto':8,'setiembre':9,'octubre':10,'noviembre':11,'diciembre':12}
```

```python
dv1['Mes_num']=dv1['Mes'].map(meses)
dv1
```

```python
dv1.info()
```

```python
vma=dv1.groupby(['Mes','Mes_num','Año'])['Importe IVA Inc. (UYU)'].sum().reset_index()
vma
```

```python
vma.sort_values((['Año','Mes_num']), inplace=True)
vma['Mes Año']=vma.groupby('Mes_num')['Importe IVA Inc. (UYU)'].pct_change()
vma['Año Año'] = vma.groupby('Año')['Importe IVA Inc. (UYU)'].pct_change()
vma
```

```python
vma.reset_index(drop=True)
```

```python
dvf=dv1.merge(vma, on=['Mes_num','Año','Mes'], how='left', suffixes=(' 1', ' Total Mes'))
dvf
```

```python
dvf.to_csv('Datos Ventas I Procesado.csv', index=False)
```
